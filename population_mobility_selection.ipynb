### Import packages.

```
import numpy as np
import pandas as pd
import json
```

### Import the data

```
lstm = pd.read_csv('/content/drive/My Drive/or_data/df_rmse_prediction_LSTM.csv',index_col=0)

bilstm = pd.read_csv('/content/drive/My Drive/or_data/df_rmse_prediction_BiLSTM.csv',index_col=0)

arima110 = pd.read_csv('/content/drive/My Drive/or_data/df_rmse_prediction_ARIMA110.csv',index_col=0)

prophet = pd.read_csv('/content/drive/My Drive/or_data/df_rmse_prediction_Prophet.csv',index_col=0)
```

```
all_performances = pd.concat([lstm[['rmse_LSTM']], bilstm[['rmse_BiLSTM']],arima110[['rmse_ARIMA110']],prophet[['rmse_Prophet']]], axis=1)
```

```
all_performances
```

```
all_performances['model_selection']=all_performances.idxmin(axis=1).values
```

```
all_performances['model_selection'][0].split('_')[1]
```

```
all_performances['model_selection'] = all_performances['model_selection'].apply(lambda x: x.split('_')[1])
```

```
all_performances
```

```
lstm_lists = list(all_performances[all_performances['model_selection'] == 'LSTM'].index)

bilstm_lists = list(all_performances[all_performances['model_selection'] == 'BiLSTM'].index)

prophet_lists = list(all_performances[all_performances['model_selection'] == 'Prophet'].index)
```

```
column_names = list(all_performances.index)

final_prediction = pd.DataFrame(columns = column_names)
```

```
prediction_LSTM = lstm.loc[lstm_lists]['prediction_LSTM']

prediction_BiLSTM = bilstm.loc[bilstm_lists]['prediction_BiLSTM']

prediction_Prophet = prophet.loc[prophet_lists]['prediction_Prophet']
```

```
for i in range(len(lstm_lists)):
  final_prediction[lstm_lists[i]] = json.loads(prediction_LSTM[i])

for i in range(len(bilstm_lists)):
  final_prediction[bilstm_lists[i]] = json.loads(prediction_BiLSTM[i])

for i in range(len(prophet_lists)):
  split_str = prediction_Prophet[i].split()
  split_str[0] = split_str[0][1:]
  split_str[-1] = split_str[-1][:-1]
  split_list = [float(element) for element in split_str]
  final_prediction[prophet_lists[i]] = split_list
```
